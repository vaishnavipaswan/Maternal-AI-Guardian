## Data preprocessing
Handle missing values, encode categorical features, and scale numerical features as necessary to prepare the data for model training.

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
import pandas as pd

# 1. Fill the missing values in the PossibleDiseases column
df['PossibleDiseases'] = df['PossibleDiseases'].fillna('No Diagnosis')

# 2. Identify and encode the categorical column RiskLevelBinary
# Check unique values in RiskLevelBinary
print("Unique values in RiskLevelBinary:", df['RiskLevelBinary'].unique())

# Since there are two unique values, use Label Encoding
le = LabelEncoder()
df['RiskLevelBinary_Encoded'] = le.fit_transform(df['RiskLevelBinary'])

# 3. Identify the numerical columns that need scaling. Exclude the encoded target variable and other non-feature columns.
# Original numerical columns: 'Age', 'HeartRate', 'BodyTemp', 'BS', 'SystolicBP', 'DiastolicBP', 'StressLevel', 'SleepQuality'
# Binary columns that don't need scaling: 'Swelling', 'Nausea', 'AbdominalPain', 'VaginalBleeding', 'DepressionSymptoms', 'Diabetes', 'Anemia', 'PriorMiscarriage', 'RiskLevelBinary_Encoded'
numerical_cols = ['Age', 'HeartRate', 'BodyTemp', 'BS', 'SystolicBP', 'DiastolicBP', 'StressLevel', 'SleepQuality']

# 4. Apply a standard scaler to the identified numerical columns
scaler = StandardScaler()
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

# 5. Concatenate the scaled numerical features and the encoded categorical features
# The binary columns are already in numerical form and don't need scaling, so they can be included directly.
# The 'PossibleDiseases' and original 'RiskLevelBinary' columns are not features for this model.
features = df[numerical_cols + ['Swelling', 'Nausea', 'AbdominalPain', 'VaginalBleeding', 'DepressionSymptoms', 'Diabetes', 'Anemia', 'PriorMiscarriage']]
target = df['RiskLevelBinary_Encoded']

# Display the first few rows of the processed features and target
print("\nProcessed Features (first 5 rows):")
display(features.head())
print("\nTarget Variable (first 5 rows):")
display(target.head())
```

## Feature selection
Select the relevant features for predicting risk level and possible diseases based on the initial exploration and domain knowledge.


```python
# Define the list of selected features based on relevance to maternal risk
selected_features_list = [
    'Age',
    'HeartRate',
    'BodyTemp',
    'BS',
    'SystolicBP',
    'DiastolicBP',
    'Swelling',
    'Nausea',
    'AbdominalPain',
    'VaginalBleeding',
    'StressLevel',
    'DepressionSymptoms',
    'SleepQuality',
    'Diabetes',
    'Anemia',
    'PriorMiscarriage'
]

# Create a new DataFrame with only the selected features
selected_features = features[selected_features_list]

# Display the first few rows of the selected features DataFrame
print("Selected Features for Model Training (first 5 rows):")
display(selected_features.head())

# Display the shape of the selected features DataFrame
print("\nShape of the Selected Features DataFrame (rows, columns):")
display(selected_features.shape)
```

## Data splitting
Split the dataset into training and testing sets to evaluate the model's performance on unseen data.


```python
from sklearn.model_selection import train_test_split

# Split the selected_features DataFrame and the target Series
X_train, X_test, y_train, y_test = train_test_split(
    selected_features,
    target,
    test_size=0.2,  # 20% for testing
    random_state=42  # for reproducibility
)

# Print the shapes of the training and testing sets
print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

## Model training

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier

# Instantiate the models
log_reg = LogisticRegression(random_state=42)
svc = SVC(random_state=42)
rf_clf = RandomForestClassifier(random_state=42)
gb_clf = GradientBoostingClassifier(random_state=42)
knn_clf = KNeighborsClassifier()

# Train the models
print("Training Logistic Regression...")
log_reg.fit(X_train, y_train)
```

```python
print("Training Support Vector Machine...")
svc.fit(X_train, y_train)
```

```python
print("Training Random Forest Classifier...")
rf_clf.fit(X_train, y_train)
```

## Model evaluation
Evaluate the trained models using appropriate metrics, focusing on accuracy, sensitivity, and specificity, given the healthcare context.


```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Evaluate Logistic Regression
print("Evaluating Logistic Regression:")
y_pred_log_reg = log_reg.predict(X_test)
accuracy_log_reg = accuracy_score(y_test, y_pred_log_reg)
print(f"Accuracy: {accuracy_log_reg:.4f}")
print("Classification Report:")
print(classification_report(y_test, y_pred_log_reg))
conf_matrix_log_reg = confusion_matrix(y_test, y_pred_log_reg)
print("Confusion Matrix:")
print(conf_matrix_log_reg)
# Specificity calculation: TN / (TN + FP)
# Confusion matrix structure: [[TN, FP], [FN, TP]]
tn_log_reg, fp_log_reg, fn_log_reg, tp_log_reg = conf_matrix_log_reg.ravel()
specificity_log_reg = tn_log_reg / (tn_log_reg + fp_log_reg) if (tn_log_reg + fp_log_reg) > 0 else 0
print(f"Specificity: {specificity_log_reg:.4f}")
print("-" * 30)

# Evaluate SVM
print("Evaluating Support Vector Machine:")
y_pred_svc = svc.predict(X_test)
accuracy_svc = accuracy_score(y_test, y_pred_svc)
print(f"Accuracy: {accuracy_svc:.4f}")
print("Classification Report:")
print(classification_report(y_test, y_pred_svc))
conf_matrix_svc = confusion_matrix(y_test, y_pred_svc)
print("Confusion Matrix:")
print(conf_matrix_svc)
tn_svc, fp_svc, fn_svc, tp_svc = conf_matrix_svc.ravel()
specificity_svc = tn_svc / (tn_svc + fp_svc) if (tn_svc + fp_svc) > 0 else 0
print(f"Specificity: {specificity_svc:.4f}")
print("-" * 30)

# Evaluate Random Forest
print("Evaluating Random Forest Classifier:")
y_pred_rf_clf = rf_clf.predict(X_test)
accuracy_rf_clf = accuracy_score(y_test, y_pred_rf_clf)
print(f"Accuracy: {accuracy_rf_clf:.4f}")
print("Classification Report:")
print(classification_report(y_test, y_pred_rf_clf))
conf_matrix_rf_clf = confusion_matrix(y_test, y_pred_rf_clf)
print("Confusion Matrix:")
print(conf_matrix_rf_clf)
tn_rf_clf, fp_rf_clf, fn_rf_clf, tp_rf_clf = conf_matrix_rf_clf.ravel()
specificity_rf_clf = tn_rf_clf / (tn_rf_clf + fp_rf_clf) if (tn_rf_clf + fp_rf_clf) > 0 else 0
print(f"Specificity: {specificity_rf_clf:.4f}")
print("-" * 30)

# Evaluate Gradient Boosting
print("Evaluating Gradient Boosting Classifier:")
y_pred_gb_clf = gb_clf.predict(X_test)
accuracy_gb_clf = accuracy_score(y_test, y_pred_gb_clf)
print(f"Accuracy: {accuracy_gb_clf:.4f}")
print("Classification Report:")
print(classification_report(y_test, y_pred_gb_clf))
conf_matrix_gb_clf = confusion_matrix(y_test, y_pred_gb_clf)
print("Confusion Matrix:")
print(conf_matrix_gb_clf)
tn_gb_clf, fp_gb_clf, fn_gb_clf, tp_gb_clf = conf_matrix_gb_clf.ravel()
specificity_gb_clf = tn_gb_clf / (tn_gb_clf + fp_gb_clf) if (tn_gb_clf + fp_gb_clf) > 0 else 0
print(f"Specificity: {specificity_gb_clf:.4f}")
print("-" * 30)

# Evaluate K-Nearest Neighbors
print("Evaluating K-Nearest Neighbors Classifier:")
y_pred_knn_clf = knn_clf.predict(X_test)
accuracy_knn_clf = accuracy_score(y_test, y_pred_knn_clf)
print(f"Accuracy: {accuracy_knn_clf:.4f}")
print("Classification Report:")
print(classification_report(y_test, y_pred_knn_clf))
conf_matrix_knn_clf = confusion_matrix(y_test, y_pred_knn_clf)
print("Confusion Matrix:")
print(conf_matrix_knn_clf)
tn_knn_clf, fp_knn_clf, fn_knn_clf, tp_knn_clf = conf_matrix_knn_clf.ravel()
specificity_knn_clf = tn_knn_clf / (tn_knn_clf + fp_knn_clf) if (tn_knn_clf + fp_knn_clf) > 0 else 0
print(f"Specificity: {specificity_knn_clf:.4f}")
print("-" * 30)
```

```python
# Install TensorFlow.js converter if not already installed
!pip install -q tensorflowjs

# ---- Step 1: Import Libraries ----
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import tensorflow as tf
import tensorflowjs as tfjs

# ---- Step 2: Load and Preprocess Dataset ----
df = pd.read_csv("/content/maternal_risk_with_disease_diagnosis.csv")

# Optional: Export CSV to JSON
df.to_json("maternal_risk_data.json", orient='records', indent=4)

# Fill missing
df['PossibleDiseases'] = df['PossibleDiseases'].fillna('No Diagnosis')

# Encode target
le = LabelEncoder()
df['RiskLevelBinary_Encoded'] = le.fit_transform(df['RiskLevelBinary'])

# Feature Selection
selected_features = [
    'Age', 'HeartRate', 'BodyTemp', 'BS', 'SystolicBP', 'DiastolicBP',
    'Swelling', 'Nausea', 'AbdominalPain', 'VaginalBleeding',
    'StressLevel', 'DepressionSymptoms', 'SleepQuality',
    'Diabetes', 'Anemia', 'PriorMiscarriage'
]
X = df[selected_features]
y = df['RiskLevelBinary_Encoded']

# Normalize
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train/Test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# ---- Step 3: Build & Train Keras Model ----
model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=20, validation_split=0.1)

# ---- Step 4: Save Model in .h5 Format ----
model.save("maternal_risk_model.h5")
print("✅ Saved .h5 model as maternal_risk_model.h5")

# ---- Step 5: Convert to TensorFlow.js Format ----
tfjs.converters.save_keras_model(model, "tfjs_model")
print("✅ Converted model to TensorFlow.js format")
```
